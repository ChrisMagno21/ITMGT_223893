```python
import time
import requests
import matplotlib.pyplot as plt

list_1 = []

for i in range(103):
    print(i)
    res = requests.get('http://api.open-notify.org/iss-now.json')
    json_data = res.json()
    list_1.append(json_data)
    time.sleep(10)
```

```python
x=[]
for i in range(0,103):
    latitude = list_1[i].get('iss_position',{}).get('latitude',[])
    x.append(latitude)
    
print(x)
    
y=[]
for j in range(0,103):
    longitude = list_1[j].get('iss_position',{}).get('longitude',[])
    y.append(longitude)

print(y)

plt.plot(x, y)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Position of the ISS space station')
plt.ylim(min(y), max(y))
plt.show()
```
